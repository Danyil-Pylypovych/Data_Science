# Звіт про Аналіз взаємозалежності факторів

## Вступ

У цьому звіті проведено аналіз даних по автомобілям. Для виконання цього аналізу було використана мова програмування Python. Аналіз був проведений завдяки графікам.

## Залежність числових даних між собою

![Залежність числових даних між собою](Залежність_всіх_числових_даних.png)

**Висновок 1:** На цьому графіку видно всі залежності кожного числового значення між собою. Наприклад залежності ціни від інших числових факторів. Можна побачити чітко що позитивну залежність, тобто якщо одне значення більше, то інше теж більше з наступними полями як: wheel-base, lenght, width, curb-weight, engine-size, bore, horsepower. Позитивне значення цих полів позитивно впливає на ціну автівки. Також є декілько полів які навпаки чим більше значення поля тим меньше ціна. Вони не прям дуже вирежанні, але тенденція присутня в полях city-mpg, highway-mpg. Тому ці поля впливають на ціни на автівки і це треба враховувати.

## Перевірка на викиди

![Перевірка на викиди](Графік_яшику_з_вусами.png)

**Висновок 2:** На графіку показані точки у вигляді ромба, це є можливі викиди, наголошую можливі. Від них можна позбавитись, або не чипати. Було спроба позбутись їх, але на графіку нижче показано що не від всіх можливих викидів позбулись, але було прийнято рішення що треба залишити їх. Можливих викидів небагато і вони вписуються в діапазон заданих в контексті завдання, тому вважається що вони не повині негативно повпливати на аналіз.

## Теплова карта взаємозв'язків між полями

![Теплова карта взаємозв'язків між полями](Взаємозв'язки_між_полями.png)

**Висновок 3:** Графік показує наскільки поле впливає на інше поле при значенні. Більш червоний означає що позитивно, прозоре ніяк, а сине негативно. Наприклад якщо поле curb-weight високе, то і скоріш за все поле engine-size буде вище. Теж саме для інших полів які мають вище, або приблизний до цього значення. Також є поля які негативно впливають один на одного, наприклад engine-location який має текстовий формат, і він переформатований в числовий формат. Це поле містить 2 значення front, rear. Якщо перше значення, то поле compression-ratio більше і навпаки. Так що тут можна багато чого розписати по кожному з пар полей і подивитись наскільки впливає одне значення від іншого.
