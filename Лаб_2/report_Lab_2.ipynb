# Звіт про Аналіз Нерухомості

## Вступ

У цьому звіті проведено аналіз даних з нерухомості для різних районів. Використовувався для аналізу даних мова програмування Python для побудови та аналізу даних про Манхетен, Бронкс, Бруклін, Квінс та Тейтен Айленд.

## Середня ціна по кварталам років

![Середня ціна продажу нерухомості по кварталам років](Serena_cena_prodazhu_po_kvartalam_bronx.png)

**Висновок 1:** З графіка видно, що район Манхетен сильно дорожчий за інші райони. Також можна побачити, що ціни продажу не сильно змінюються, але в Манхетені падала ціна з кінця 2022 року по середину 2023 року.

## Розподіл по категоріям

![Розподіл по категоріям](Розподіл_по_категоріямstatenisland.png)

**Висновок 2:** Аналіз показав, що комерційних будівель досить мало в порівнянні з іншими типами. Також зазначимо що в районах Бронкс та Стейтен Айланд меньше будівель ніж в інших районах, а в Манхетені в основному присутні змішаний тип побудівель.

## Вплив розміру ділянки на ціну

![Вплив розміру ділянки на ціну](Вплив_розміру_ділянки_на_ціну_statenisland.png)

**Висновок 3:** Аналіз показує, що розмір ділянки майже не впливає на ціну продажу, тому не можна буде продати дорожче опираючись тільки на розмір ділянки. В Манхетені взагалі вплив від'ємний, хоч і дуже близький нулю, тому в Манхетені скоріш побачиш дорожчими будівлі де меньші ділянки, але знов зазначу що вплив розміру ділянки на ціну близький нулю.

## Вплив року будівництва на ціну

![Вплив року будівництва на ціну](Вплив_року_будівництва_на_ціну_statenisland.png)

**Висновок 4:** З графіку видно, що вплив року будівництва також не впливає на ціну продажу. Показано що в усіх районах досить рівномірно розподілені ціни продажу по роках. В Манхетені і в Бронксі є всплески в деяких періодах по рокам цін продажу, але це не залежить від новизни чи старших побудівель, бо всплески наприклад в Манхетені в періоди 1900 по 1925 та з 2000 по наш час. Тому можна сказати, що на зараз рік побудови не впливає на ціну, але якщо ціни продажу будуть рости, то можливо зміниться це.

## Середні ціни продажу за категорією будівельного класу

![Середні ціни продажу за категорією будівельного класу](Середні_ціни_продажу_за_категорією_будівельного_класу_statenisland.png)

**Висновок 5:** З цього графіку можна сказати, що категорії будівельного класу не сильно впливають на середню ціну продажу, але можна побачити, що в Манхетені значно більша середня ціна продажу нерухомості. Також можна побачити кількість приблизна будівель певної категорії класу. Також зазначу що в Стейтен Айленд досить низька середня ціна продажу. Думаю можна сказати, що якщо краще продавати нерухомість в Манхетені, якщо цікавить більш велика ціна продажу.

## Типи оподаткування

![Типи оподаткування](Типи_оподаткування_statenisland.png)

**Висновок 6:** На графіку видно, що в усіх районах, окрім Манхетена значна і кількість типів оподаткування є 1, також і 2 досить багато. А в Манхетені тип оподаткування сильно домінує саме 2, це треба враховувати і це можеть створити труднощі, або меньший прибуток, якщо продавати нерухомість в Манхетені.

![Типи оподаткування](Вплив_розміру_будівлі_на_ціну_statenisland.png)

**Висновок 7:** Тут вже є тенденція, що розмір будівлі позитивно впливає на ціну більш меньш нормально. Але для Манхетену це всеодно досить низький показник. Але якщо брати інші райони то там вплив є і чим більше розмір тим більша ціна продажу.
