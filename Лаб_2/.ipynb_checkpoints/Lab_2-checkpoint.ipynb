```python
boroughs = ['manhattan', 'bronx', 'brooklyn', 'queens', 'statenisland']

fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

for i, borough in enumerate(boroughs):
    data = pd.read_excel(f'rollingsales_{borough}.xlsx')
    data = data.dropna(subset=["SALE PRICE", "SALE DATE"])
    data["SALE DATE"] = pd.to_datetime(data["SALE DATE"])
    data["YEAR"] = data["SALE DATE"].dt.year
    data["QUARTER"] = data["SALE DATE"].dt.quarter

    average_prices = data.groupby(["YEAR", "QUARTER"])["SALE PRICE"].mean().reset_index()

    sb.barplot(data=average_prices, x="YEAR", y="SALE PRICE", hue="QUARTER", ax=axes[i % 3])
    axes[i % 3].set_title(borough.capitalize())
    axes[i % 3].set_xlabel("Рік")
    axes[i % 3].set_ylabel("Середня ціна продажу")

    if (i + 1) % 3 == 0:
        plt.tight_layout()
        plt.show()
        fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

if (i + 1) % 3 != 0:
    plt.tight_layout()
    plt.show()
```

```python
boroughs = [manhattan, bronx, brooklyn, queens, statenisland]

# Додавання категорії до кожного DataFrame
for df in boroughs:
    df["CATEGORY"] = "Комбіновані"
    df.loc[df["COMMERCIAL UNITS"] == 0, "CATEGORY"] = "Житлові"
    df.loc[df["RESIDENTIAL UNITS"] == 0, "CATEGORY"] = "Комерційні"

# Створення фігури з підграфіками
num_rows = 2
num_cols = 3  # Максимум по 3 графіка в ряд

fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 6 * num_rows), sharey=True)

for i, df in enumerate(boroughs):
    combined_data = df[df["CATEGORY"] == "Комбіновані"]
    residential_data = df[df["CATEGORY"] == "Житлові"]
    commercial_data = df[df["CATEGORY"] == "Комерційні"]

    # Отримання поточного підграфіка
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]

    # Побудова графіка для поточного району
    ax.bar(combined_data["CATEGORY"], combined_data["CATEGORY"].count(), width=0.4, label="Комбіновані", align="center")
    ax.bar(residential_data["CATEGORY"], residential_data["CATEGORY"].count(), width=0.4, label="Житлові", align="edge", color="green")
    ax.bar(commercial_data["CATEGORY"], commercial_data["CATEGORY"].count(), width=0.4, label="Комерційні", align="edge", color="red")
    ax.set_title(f"Розподіл об'єктів у категоріях {df}")
    ax.set_xlabel("Категорія")
    ax.set_ylabel("Кількість об'єктів")
    ax.legend()

# Завершення побудови графіків
plt.tight_layout()
plt.show()
```
